EPSG:5186

```python
df_wel = pd.read_csv('./df_wel_prepo_EPSG_4326.csv')
```

```python
df_wel = df_wel[['시설명', '시설주소', 'X', 'Y']]

# 좌표 변환 함수 정의
def transform_coordinates(x, y):
    transformer = pyproj.Transformer.from_crs("EPSG:4326", "EPSG:5186", always_xy=True)
    new_x, new_y = transformer.transform(x, y)
    return new_x, new_y

# DataFrame의 각 행에 변환 함수 적용
df_wel['X'], df_wel['Y'] = zip(*df_wel.apply(lambda row: transform_coordinates(row['X'], row['Y']), axis=1))
```

```python
df_school = pd.read_csv('./서울시_특수학교_기본정보_변환.csv')
```

```python
df_school = df_school.rename(columns={'경도': 'X', '위도': 'Y', '학교명': '시설명', '도로명주소':'시설주소'})
df_school = df_school[['시설명', '시설주소', 'X', 'Y']]
df_school
```

```python
df_welfare = pd.concat([df_wel, df_school])
df_welfare = df_welfare.reset_index(drop=True)
df_welfare
```

```python
df_welfare.to_csv('welfare_school_concat.csv', index=False)
```
