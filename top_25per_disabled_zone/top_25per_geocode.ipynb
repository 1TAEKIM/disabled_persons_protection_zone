```python
missing_latitude_rows = df[df['Latitude'].isnull()]
print(missing_latitude_rows)
```

```python
import folium

# 결측값이 없는 행만 추출
df_with_coordinates = df.dropna(subset=['Latitude', 'Longitude'])

# 지도 초기화
map_center = [df_with_coordinates['Latitude'].mean(), df_with_coordinates['Longitude'].mean()]
mymap = folium.Map(location=map_center, zoom_start=12)

# 각 행에 대한 마커 추가
for _, row in df_with_coordinates.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['시설명'],  # 팝업에 보여줄 정보 선택
        icon=folium.Icon(color='blue')
    ).add_to(mymap)

# 지도를 HTML 파일로 저장
# mymap.save('map.html')
mymap
```
